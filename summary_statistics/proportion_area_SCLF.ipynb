```python
%%time
import xarray as xr
import rioxarray
import geopandas as gpd

# Load the TIFF file as an xarray dataset
tif_path = "SCLF_model.tif"
raster = rioxarray.open_rasterio(tif_path, masked=True)

# Load the polygons as a GeoDataFrame
gdf = gpd.read_file("koppen_EPSG3577_arcgis_clip.shp")
# gdf = gpd.read_file("STE_2021_AUS_EPSG3577_TAS.gpkg")

# Ensure CRS matches between raster and polygons
gdf = gdf.to_crs(raster.rio.crs)

# Define the intervals
intervals = [(i / 10, (i + 1) / 10) for i in range(10)]  # [(0.0, 0.1), (0.1, 0.2), ..., (0.9, 1.0)]

# Initialize columns for each interval
for lower, upper in intervals:
    column_name = f"area_{lower:.1f}_{upper:.1f}_km2"  # Updated column names to reflect km^2
    gdf[column_name] = 0.0  # Initialize with 0.0 for area in km^2

# Iterate over each polygon
for idx, polygon in gdf.iterrows():
    print(polygon)
    # Clip the raster to the polygon
    clipped = raster.rio.clip([polygon.geometry], from_disk=True)
    
    # Count pixels for each interval
    for lower, upper in intervals:
        count = ((clipped > lower) & (clipped <= upper)).sum().item()
        area_km2 = count * 0.0009  # Convert pixel count to km^2
        column_name = f"area_{lower:.1f}_{upper:.1f}_km2"
        gdf.at[idx, column_name] = area_km2  # Update the corresponding column

# Save the updated GeoDataFrame
gdf.to_file("koppen_EPSG3577_arcgis_clip_SCLF_model.gpkg")
# gdf.to_file("STE_2021_AUS_EPSG3577_TAS_SCLF_km2.gpkg")
```
